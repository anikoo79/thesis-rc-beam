```python
import os, sys, json, time, random, math, platform
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---- Paths (relative to project root) ----
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
DATA_DIR     = os.path.join(PROJECT_ROOT, "data")
REPORT_DIR   = os.path.join(PROJECT_ROOT, "reports")
FIG_DIR      = os.path.join(REPORT_DIR, "figures")
TABLE_DIR    = os.path.join(REPORT_DIR, "tables")
for d in [DATA_DIR, REPORT_DIR, FIG_DIR, TABLE_DIR]:
    os.makedirs(d, exist_ok=True)

# ---- Reproducibility ----
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# ---- Plot defaults ----
plt.rcParams["figure.figsize"] = (6, 4)
plt.rcParams["axes.grid"] = True

print("Python:", sys.version.split()[0], "| OS:", platform.platform())
print("Project root:", PROJECT_ROOT)
print("Dirs:", {"data": DATA_DIR, "figures": FIG_DIR, "tables": TABLE_DIR})

# ---- Sanity I/O test: write & read a tiny table ----
df = pd.DataFrame({"x": np.arange(5), "y": np.arange(5) ** 2})
test_csv = os.path.join(DATA_DIR, "sanity_table.csv")
df.to_csv(test_csv, index=False)
print("Wrote:", test_csv)

df2 = pd.read_csv(test_csv)
print("Read shape:", df2.shape)

# ---- Quick plot (saved to reports/figures) ----
plt.plot(df2["x"], df2["y"])
plt.title("Sanity Plot: y = x^2")
plt.tight_layout()
out_png = os.path.join(FIG_DIR, "sanity_plot.png")
plt.savefig(out_png, dpi=150)
plt.show()
print("Saved figure:", out_png)
```
